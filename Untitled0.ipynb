```python
import fitz  # PyMuPDF
from langchain_text_splitters import RecursiveCharacterTextSplitter
from sentence_transformers import SentenceTransformer
import chromadb

# 1. Extract Text
doc = fitz.open("General Policies and Procedures 1.pdf") # Replace with your filename
text = "".join([page.get_text() for page in doc])

# 2. Chunking (Breaking text into searchable pieces)
splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
chunks = splitter.split_text(text)

# 3. Load Embedding Model
# This downloads a ~100MB model to Colab's RAM
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# Initialize In-Memory Client
client = chromadb.Client()
collection = client.create_collection(name="pdf_search")

# Convert chunks to embeddings and add to DB
embeddings = model.encode(chunks).tolist()
collection.add(
    documents=chunks,
    embeddings=embeddings,
    ids=[f"id_{i}" for i in range(len(chunks))]
)
```

```python
query = "What does the document say about [Offshore Development Centers]?"
query_embedding = model.encode([query]).tolist()

results = collection.query(query_embeddings=query_embedding, n_results=2)
print("Best Match:", results['documents'][0][0])
```
